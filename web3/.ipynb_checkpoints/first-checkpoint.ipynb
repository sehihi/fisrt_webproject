```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 맑은 고딕으로 설정

# 음수 기호가 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 엑셀 파일 읽기
file_path = "1_호선별 개정도 현황_호선별 개정도 LIST(3개 호선)_설명포함.xlsx"
df = pd.read_excel(file_path, sheet_name="P1 개정도")

# 출도부서와 원인코드 간의 교차 테이블 생성
crosstab = pd.crosstab(df['출도부서'], df['원인코드'])

# 히트맵으로 시각화
plt.figure(figsize=(10, 8))

# imshow로 히트맵 표시
heatmap = plt.imshow(crosstab, cmap="YlGnBu", aspect='auto')

# 컬럼과 인덱스에 라벨 붙이기
plt.xticks(np.arange(len(crosstab.columns)), crosstab.columns, rotation=45, ha="right", fontsize=10)
plt.yticks(np.arange(len(crosstab.index)), crosstab.index, fontsize=10)

# 컬러바 추가
cbar = plt.colorbar(heatmap)
cbar.set_label("Frequency", rotation=270, labelpad=20, fontsize=12)

# 그래프 제목 추가
plt.title("출도부서와 원인코드 간의 관계", fontsize=16, pad=20)

# 각 셀에 값 추가 (선택 사항)
for i in range(len(crosstab.index)):
    for j in range(len(crosstab.columns)):
        plt.text(j, i, crosstab.iloc[i, j], ha="center", va="center", color="black", fontsize=9)

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일 읽기
file_path = "1_호선별 개정도 현황_호선별 개정도 LIST(3개 호선)_설명포함.xlsx"
df = pd.read_excel(file_path, sheet_name="P1 개정도")

# 첫 번째 컬럼('출도부서')의 첫 번째 행의 값에 해당하는 데이터 필터링
selected_value = df['출도부서'].iloc[0]
filtered_df = df[df['출도부서'] == selected_value]

# 두 번째 컬럼('원인코드')의 값들에 대한 빈도 계산
count_by_cause = filtered_df['원인코드'].value_counts()

# 막대그래프로 시각화
plt.figure(figsize=(10, 6))
count_by_cause.plot(kind='bar', color='skyblue')

# 그래프 제목 및 레이블 설정
plt.title(f"출도부서: {selected_value}와 원인코드 간의 관계", fontsize=16)
plt.xlabel('원인코드', fontsize=14)
plt.ylabel('빈도', fontsize=14)

# 그래프 출력
plt.show()
```

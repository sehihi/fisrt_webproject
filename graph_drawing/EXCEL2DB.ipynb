```python
import pymongo
from openpyxl import load_workbook

def Connect_DB():
    connect_to = pymongo.MongoClient("localhost", 3000)
    mdb = connect_to.graph
    collection = mdb.v_data
    return collection

def excel_to_DB(collection):
    wb = load_workbook("5.xlsx")
    ws = wb.active
    for x in range(1, ws.max_row+1):
        db_data = {
            "프로젝트": ws.cell(row=x, column=1).value,
            "선종": ws.cell(row=x, column=2).value,
            "선종2": ws.cell(row=x, column=3).value,
            "업체구분": ws.cell(row=x, column=4).value,
            "검사길이": ws.cell(row=x, column=5).value,
            "불량길이": ws.cell(row=x, column=6).value,
            "검사방법": ws.cell(row=x, column=7).value,
            "검사자ID": ws.cell(row=x, column=8).value,
            "과": ws.cell(row=x, column=9).value,
            "검사실적입력일": ws.cell(row=x, column=10).value,
            "NDE업체": ws.cell(row=x, column=11).value,
            "용도판정": ws.cell(row=x, column=12).value,
            "사유코드": ws.cell(row=x, column=13).value,
            "사유코드 설명": ws.cell(row=x, column=14).value,
            "직반장사번": ws.cell(row=x, column=15).value,
            "직반장성명": ws.cell(row=x, column=16).value,
            "직반장 Vendor 코드": ws.cell(row=x, column=17).value,
            "용접사사번": ws.cell(row=x, column=18).value,
            "용접사성명": ws.cell(row=x, column=19).value,
            "용접완료일": ws.cell(row=x, column=20).value,
            "용접방법": ws.cell(row=x, column=21).value
        }
        collection.insert_one(db_data)

collection = Connect_DB()
excel_to_DB(collection)
```
